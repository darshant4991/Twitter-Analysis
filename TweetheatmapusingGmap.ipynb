```python
import os
import tweepy as tw
import pandas as pd
import gmaps
import gmaps.datasets
tw_key = 'ZMWOLa'
tw_secret = 'CrJ061ey'
tw_access_token = '1499373100'
tw_access_token_secret = '97mH720sQp1d'
auth = tw.OAuthHandler(tw_key, tw_secret)
auth.set_access_token(tw_access_token, tw_access_token_secret)
api = tw.API(auth, wait_on_rate_limit=True)
search_words = "Ministry of Truth"
```

```python
tweets = tw.Cursor(api.search, q=search_words, lang="en").items(2000)
MOTTweets = [[tweet.id,tweet.user.screen_name, tweet.user.verified, tweet.text, tweet.place, tweet.user.location, tweet.source,tweet.created_at,tweet.retweet_count,tweet.favorite_count] 
             for tweet in tweets]
MOTTweetsdf = pd.DataFrame(data=MOTTweets, columns=['id','user',"verified","text","source","place","TweetSource","created_at","retweetCount","Likes"])
MOTTweetsdf.to_csv(r'extracted_Tweets_with_IDtry.csv', index = None, header=True) 
```

```python
from geopy.geocoders import Nominatim
MOTTweets_locations = {"lat": [], "long": []}
geolocator = Nominatim(user_agent="presentation")
for count, user_loc in enumerate(MOTTweetsdf['place']):
  try:
    location = geolocator.geocode(user_loc)
    if location:
      MOTTweets_locations['lat'].append(location.latitude)
      MOTTweets_locations['long'].append(location.longitude)
  except:
    pass
MOTTweets_locations = pd.DataFrame(MOTTweets_locations).to_csv(r"MOTtweet_locationstry.csv", index = None, header=True)
```
